```python
from csv import excel


def key_id(value):
    value = str(value).lower()
    if 'scrum' in value or 'csm' in value or 'smum' in value or 'srum' in value:
        return 1
    elif 'amazon' in value or 'aws' in value or 'devops' in value:
        return 2
    elif 'pmp' in value or 'project management' in value or 'pmi' in value or 'proyectos' in value or 'it project' in value:
        return 3
    elif 'cloud' in value:
        return 4
    elif 'capm' in value:
        return 5
    elif 'cspo' in value:
        return 6
    elif 'itil' in value or 'itl' in value:
        return 7
    elif 'simpl' in value or 'simli' in value or 'simpi' in value or 'smpli' in value or 'simi' in value or 'sipli' in value:
        return 8
    elif 'safe' in value or 'scale' in value:
        return 9
    elif 'togaf' in value or 'udacity' in value or 'it architect' in value:
        return 10
    else:
        return None
    
    
    
df['Keyword ID'] = df['Keyword'].apply(key_id)
df.head(10)
```

```python
# Apply the key_id function to assign each row a Keyword ID
df['Keyword ID'] = df['Keyword'].apply(key_id)
```

```python
# Drop the 'Trends' column if present
if 'Trends' in df.columns:
    df = df.drop('Trends', axis=1)

# Reorder columns to match website_traffic_data format
cols = [
    'Title', 'Keyword', 'Keyword ID', 'Position', 'Previous position', 
    'Last Seen', 'Search Volume', 'CPC', 'Traffic', 'Traffic (%)', 
    'Traffic Cost', 'Traffic Cost (%)', 'Competition', 'Number of Results', 
    'Keyword Difficulty'
]
df = df[cols]

df.head(10)
```

```python
# Define the mapping keywords
keyword_map = {
    "Scrum Master": ['scrum', 'csm', 'smum', 'srum', 'scrummaster'],
    "AWS": ['amazon', 'aws','devops'],
    "Cloud": ['cloud'],
    "PMP": ['pmp', 'project management', 'pmi', 'proyectos', 'it project'],
    "CAPM": ['capm'],
    "CSPO": ['cspo'],
    "ITIL": ['itil', 'itl'],
    "Simplilearn": ['simplilearn', 'simpli', 'simpi', 'smpli', 'simi', 'sipli', 'simply','simlilearn','simplelern',
                    'simplelearn', 'simplylearn', 'simplileanr','simpllearn','simple','simplearn',],
    "SAFe": ['safe', 'scaled agile', 'scale agile'],
    "TOGAF": ['togaf', 'udacity', 'it architect'],
}

# Function to categorize each row
def classify_keyword(value):
    val = str(value).lower()
    for label, keywords in keyword_map.items():
        if any(k in val for k in keywords):
            return label
    return None  # Or "Other" if you prefer

# Apply to your dataframe
df['Keyword'] = df['Keyword'].apply(classify_keyword)

# Show sample
df.head(10)
```

```python
# Save the main table to CSV
df.to_csv('website_traffic_data.csv', index=False)
```

```python
# Create a lookup table of unique Keyword IDs
keyword_table = df[['Keyword ID','Keyword']].drop_duplicates().sort_values(by='Keyword ID')
keyword_table.to_csv('keyword.csv', index=False)
```
